# Import ClassLabel from Huggingface datasets library



```python
from datasets import load_dataset, ClassLabel
```

# Load example dataset to have something to work with


```python
dataset = load_dataset("beans")
```

## There is already a 'labels' feature in this example dataset, but we'll add a new feature called 'label' as an example

```python
dataset['train'].features
```

# Write out the list of class labels that you want
(this is just an example)

```python
class_names = ['email', 'presentation', 'scientific report']
```

#Make a new ClassLabel feature using our class_names and add it as 'label' to our existing features

```python
dataset['train'].features['label'] = ClassLabel(names=class_names)
```
